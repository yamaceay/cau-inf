Non-uniqueness of graph structures

For every joint distribution $p(x, y)$ of two real-valued variables, there is a SCM $x := \eta_X, Y := f_Y(X, \eta_Y)$ with measurable $f_Y$ and independent noise variables $\eta_X, \eta_Y$

Sketch of proof

Define $f_Y(x, \eta_Y) := F_{Y | X}^{-1} (\eta_Y)$

where $F_{Y | x}(y) = P(Y \le y | X = x)$ is the conditional cumulative distribution function. Draw $\eta_Y$ uniformly and independent of $X$.

The way out: Consider restricted classes of SCMs with restrictions on the f_i

Definition: Linear additive models

The joint distribution $p(x, y)$ of two continuous random variables $X$ and $Y$ is said to satisfy a linear additive noise model from X to Y if it is generated by an SCM of the form $X := \eta_X, Y := \alpha X + \eta_Y$ with $\eta_X \perp\!\!\!\perp \eta_Y$

Theorem: Identifiability of linear non-Gaussian additive noise models

If $p(X, Y)$ admits a linear additive model from $X$ to $Y$, then it also admits a linear additive noise model from $Y$ to $X$ if and only if both $X$ and $\eta_Y$ are Gaussian.

Assume that $p(x, y)$ admits a linear non-Gaussian additive noise model from $X$ to $Y$ or vice versa

To decide which is the case, apply the following method:

1. Linearly regress $X$ on $Y$, giving rise to the predictor $\hat{Y} (X)$

2. Test whether residual $r_Y = Y - \hat{Y}(X)$ is independent of $X$

3. Perform the previous steps with $X$ and $Y$ swapped

4. If independence is accepted in one direction and not in the other, say $r_Y \perp\!\!\!\perp X$ but $r_X \not\perp\!\!\!\perp Y$, then infer $X \to Y$

Theorem: Darmois-Skitovic

Let $X_1, …, X_n$ be independent and non-degenerate random variables. If there are non-zero numbers $a_1, …, a_n$ and $b_1, …, b_n$ such that:

$I_1 = a_1X_1 + … + a_nX_n$ and $I_2 = b_1X_1 + … + b_nX_n$

are independent random variables, then all $X_i$ are Gaussian

Sketch for proof of identifiability

Assume there is a LiNGAM in both directions, i.e. $Y = \alpha X + \eta_Y$ and $X = \beta Y + \eta_X$ with $X \perp\!\!\!\perp \eta_Y$ and $Y \perp\!\!\!\perp \eta_X$. Then, the above theorem applies to $I_1 = Y = \alpha X + \eta_X$ and $I_2 = \eta_X = (1 - \alpha \beta) X - \beta \eta_Y$ such that both $X$ and $\eta_Y$, and hence also $Y$ and $\eta_Y$ are Gaussian.

Abstraction of the previous findings

1. We assumed that $p(x, y)$ was generated by an SCM with $X \to Y$ that belongs to a restricted class (here: the class of linear non-gaussian additive noise models)

2. We know that $p(x, y)$ is also generated by some other SCM in which the cause and effect are interchanged, i.e. with $Y \to X$

3. However, this second SCM cannot belong to the same restricted class, it is more complicated

4. This allowed to distinguish whether $X \to Y$ or $Y \to X$

Principle of independent mechanisms

The causal generative process of a system's variables is composed of autonomous modules that do not inform or influence each other.

In the probabilistic case, this means that the conditional distribution of each variable given its causes (i.e. its mechanism) does not inform or influence the other conditional distributions.

In case we have only two variables, this reduces to an independence between the cause distribution and the mechanism producing the effect distribution. 

Why should this be a valid approach?

Informal argument based on the principle of independent mechanisms.

* The functional assignments $X_i := f(PA_i, \eta_i)$ of individual variables correspond to autonomous physical mechanisms

* Say $X \to Y$. Then $p_X$ and $p_{Y | X}$ should contain no information about each other. 

* Assume there are $l$ many different $p_X$ and $k$ many different $p_{Y | X}$, such that are $k l$ many different $p_{X, Y}$

* Unless $p_X$ and $p_{Y | X}$ are tuned to each other, which we do not expect due to second point, there are then $k l$ many $p_Y$ and $p_{X | Y}$

* This means that $p_Y$ and $p_{X | Y}$ (a-causal factorization) are more complicated than $p_X$ and $p_{Y | X}$ (causal factorization)

* We cannot apply the same argument with $X$ and $Y$ swapped because when $X \to Y$ we do not assume that $p_Y$ and $p_{X | Y}$ do not contain information about each other.

J. Pearl classifies knowledge and questions by a three-level hierarchy according to the degree of causal information that they contain:

1. Association
2. Intervention
3. Counterfactuals
